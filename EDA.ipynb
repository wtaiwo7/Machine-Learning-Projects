```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df_012 = pd.read_csv('/Users/winnietaiwo/Desktop/diabetes_012_health_indicators_BRFSS2015.csv')
df_binary_5050 = pd.read_csv('/Users/winnietaiwo/Desktop/diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
df_binary = pd.read_csv('/Users/winnietaiwo/Desktop/diabetes_binary_health_indicators_BRFSS2015.csv')

# Define a function for basic EDA
def basic_eda(df, title):
    print(f"Exploratory Data Analysis for: {title}")
    # Display the first few rows
    display(df.head())
    
    # Summary statistics
    display(df.describe())
    
    # Check for missing values
    print("Missing values per column:")
    display(df.isnull().sum())
    
    # Check for duplicate rows
    print("Number of duplicate rows:")
    print(df.duplicated().sum())
    
    # Distribution of the target variable
    print("Distribution of the target variable:")
    df[df.columns[0]].value_counts().plot(kind='bar')
    plt.show()

# Perform EDA on each dataset
basic_eda(df_012, "Diabetes 012 Health Indicators")
basic_eda(df_binary_5050, "Diabetes Binary 50/50 Split Health Indicators")
basic_eda(df_binary, "Diabetes Binary Health Indicators")

# Additional EDA: Visualizing distributions of key features
# Adjust the features as per your requirement
key_features = ['BMI', 'Age', 'Income', 'Education']

for feature in key_features:
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    sns.histplot(df_012[feature], kde=True)
    plt.title(f'{feature} distribution in Diabetes 012')

    plt.subplot(1, 3, 2)
    sns.histplot(df_binary_5050[feature], kde=True)
    plt.title(f'{feature} distribution in Diabetes Binary 50/50')

    plt.subplot(1, 3, 3)
    sns.histplot(df_binary[feature], kde=True)
    plt.title(f'{feature} distribution in Diabetes Binary')

    plt.tight_layout()
    plt.show()
```
